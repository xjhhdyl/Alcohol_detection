```python
import thinkdsp


abcde_wave = thinkdsp.read_wave('../data/getWav20.wav')
abcde_wave.play('../data/getWav20.wav')
abcde_wave.plot()
```

```python
abcde_spectrum = abcde_wave.make_spectrum()
abcde_spectrum.plot()
```

```python
abcde_spectrum.low_pass(cutoff=1000,factor=0.001)
low_wave = abcde_spectrum.make_wave()
low_wave.plot()
```

```python
low_wave.play('../data/out_getWav20.wav')
abcde_spectrum.plot()
```
