```python
import utils.thinkdsp as thinkdsp

file_name = 'SingleWav7.wav'
aduiorecord_wave = thinkdsp.read_wave('../CollectedData/WAV/'+file_name)
aduiorecord_wave.plot()
```

```python
aduiorecord_spectrum = aduiorecord_wave.make_spectrum()
aduiorecord_spectrum.plot()
```

```python
aduiorecord_spectrum.low_pass(cutoff=10000,factor=0.001)
low_wave = aduiorecord_spectrum.make_wave()
low_wave.plot()
```

```python
low_wave.play('../data/out_'+file_name)
aduiorecord_spectrum.plot()
```
